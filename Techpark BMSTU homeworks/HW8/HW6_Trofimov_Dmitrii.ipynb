Подсчитай кол-во слов в документе "War and Peace by Leo Tolstoy (ru).txt"

```
# код
with open('War and Peace by Leo Tolstoy (ru).txt', 'rb') as f:
  text = f.read().decode("cp1251").encode("utf-8").decode("utf-8")
  new_file = open("new_file.txt", "w")
  new_file.write(text)
  new_file.close()
```

```
chars_to_remove = ",.–"
def remove_in_string(new_string):
  string = new_string
  for character in chars_to_remove:
    string = string.replace(character, "")
  return string
```

```
with open("new_file.txt", "r") as input, open("without_delimeters_last.txt", "w") as output:
  lines = input.readlines()
  for line in lines:
    print(line)
    line_without = remove_in_string(line)
    print(line_without)
    try:
      output.write(line_without)
    except Exception:
      pass
```
